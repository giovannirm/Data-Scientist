```python
def NumeroBinario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    if not isinstance(numero, int) or numero < 0:
      return "Debe ingresar un número entero o mayor que se cero"

    if numero == 0:
      return numero

    numero_binario = ""
    while numero > 0:
        resto = numero % 2
        numero_binario += str(resto)
        numero = numero // 2
    return int(numero_binario[::-1])

    # if numero == 0:
    #   return ''

    # cociente, residuo = divmod(numero, 2)
    # print(cociente)
    # print(residuo)
    # return NumeroBinario(cociente) + str(residuo)
```

2) Convertir de decimal a binario las fracciones 1/2, 1/3, 1/4, 1/5, 1/6, 1/7, 1/8, 1/9. Luego analizar los resultados y observar qué particularidad se encuentra en los mismos. Se puede usar Python o una calculadora, lo importante es ver si hay algo que podemos notar...

Salida esperada:
* Fraccion 1 / 2 :  0.5  y En binario:  0.1
* Fraccion 1 / 3 :  0.3333333333333333  y En binario:  0.010101010101010101010101
* Fraccion 1 / 4 :  0.25  y En binario:  0.01
* Fraccion 1 / 5 :  0.2  y En binario:  0.001100110011001100110011
* Fraccion 1 / 6 :  0.16666666666666666  y En binario:  0.001010101010101010101010
* Fraccion 1 / 7 :  0.14285714285714285  y En binario:  0.001001001001001001001001
* Fraccion 1 / 8 :  0.125  y En binario:  0.001
* Fraccion 1 / 9 :  0.1111111111111111  y En binario:  0.000111000111000111000111

```python
import math
def FraccionDecimalaBinario(numero):

    parte_decimal, parte_entera = math.modf(numero)

    parte_entera_binario = NumeroBinario(int(parte_entera))
    parte_decimal_binario = ''

    numero_decimal = parte_decimal
    limite = 24
    continuar = True

    while continuar:
        numero_decimal = numero_decimal * 2
        decimal, entero = math.modf(numero_decimal)
        numero_decimal = decimal

        parte_decimal_binario += str(int(entero))
        if decimal == 0.0 or limite == 1:
            continuar = False
            
        limite -= 1

    return float(str(parte_entera_binario) + '.' + parte_decimal_binario)
```

```python
FraccionDecimalaBinario(1/9)
```

```python
for i in range(2, 10):
    message = "Fracción 1/{} : {} y en binario: {}".format(i, 1/i, FraccionDecimalaBinario(1/i))
    print(message)
```
